```python
def gradient_descent(x,y):
    w0_curr = 0 # Set the value of w0
    w1_curr = 45 # Set the value of w1
    iterations = 10000 # The number of iterations
    n=len(x) # N the length of the array
    learning_rate = 0.003 # Let this be our learning rate

    for i in range(iterations):

        y_predicted = w0_curr *x + w1_curr
        error_term = (1/n) * sum([val**2 for val in (y-y_predicted)])# Error Term
        w0d = -(2/n) * sum (x*(y-y_predicted)) # Differential of w0 
        w1d = -(2/n) * sum(y-y_predicted) # Differential of w1
        w0_curr = w0_curr - learning_rate * w0d
        w1_curr = w1_curr - learning_rate * w1d
        print ("w0 {}, w1 {}, error_term {} iteration {}".format(w0_curr,w1_curr,error_term,i))
        
x=np.array([0.5,1,1.5,2,2.5,3,3.5,4,4.5])

y=np.array([79.4,46.33,44.43,44.85,48.54,49.64,52.44,53,54.1]) 

gradient_descent(x,y)
```

```python
#import libraries
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#reading data 
dataset = pd.read_csv('assign2.csv')
print(dataset.shape)
dataset.head()
```

```python
#initializing our inputs and outputs
X = dataset['Drilling depth[mm]'].values
Y = dataset['Min Pressure [MPa]'].values

#mean of our inputs and outputs
x_mean = np.mean(X)
y_mean = np.mean(Y)

#total number of values
n = len(X)

#using the formula to calculate the b1 and b0
numerator = 0
denominator = 0
for i in range(n):
    numerator += (X[i] - x_mean) * (Y[i] - y_mean)
    denominator += (X[i] - x_mean) ** 2
    
b1 = numerator / denominator
b0 = y_mean - (b1 * x_mean)

#printing the coefficient
print(b1, b0)
```

```python
#plotting values 
x_max = np.max(X) + 100
x_min = np.min(X) - 100

#calculating line values of x and y
x = np.linspace(x_min, x_max, 1000)
y = b0 + b1 * x

#plotting line 
plt.plot(x, y, color='#00ff00', label='Linear Regression')

#plot the data point
plt.scatter(X, Y, color='#ff0000', label='Data Point')

# x-axis label
plt.xlabel('Drilling depth[mm]')

#y-axis label
plt.ylabel('Min Pressure [MPa]')

plt.legend()
plt.show()
```

```python
rmse = 0
for i in range(n):
    y_pred=  b0 + b1* X[i]
    rmse += (Y[i] - y_pred) ** 2
    
rmse = np.sqrt(rmse/n)

print(rmse)
```
